```python
import numpy as np
import pandas as pd
pd.set_option("display.max_rows", None)
import matplotlib as mpl 
import matplotlib.pyplot as plt
import yfinance as yf

```

```python
gcf = yf.Ticker("GC=F")
tsla = yf.Ticker("TSLA")
```

```python
oro = gcf.history(start="2010-07-01", end="2024-05-12")
tesla= tsla.history(start="2010-07-01", end="2024-05-12")

oro=oro["Close"]
tesla= tesla["Close"]
```

```python
#ajustamos los índices
```

```python
oro.index = [str(i)[0:10] for i in oro.index]
tesla.index =  [str(i)[0:10] for i in tesla.index]
```

```python
assets = pd.concat([tesla,oro], axis=1)
assets.columns= ["tesla","oro"]
```

```python
assets
```

```python
oro_nans =  assets.oro.isnull().values
tesla_nans= assets.tesla.isnull().values
oro_nans= len(oro_nans[oro_nans==True])
tesla_nans =  len(tesla_nans[tesla_nans==True])
{"indeterminados de Oro": oro_nans, "indeterminados de Tesla":tesla_nans , "porcentaje indeterminados oro":oro_nans/len(assets.oro)
, "porcentaje de indeterminados Tesla": tesla_nans/len(assets.tesla)}
```

```python
assets.tesla= assets.tesla.fillna(np.mean(assets.tesla))
assets.oro= assets.oro.fillna(np.mean(assets.oro))
```

```python
assets= assets.assign(lag_oro= assets.oro.shift(1)  ,lag_tesla = assets.tesla.shift(1))
assets = assets[["oro","lag_oro","tesla", "lag_tesla"]]
assets
```

### calculamos la variación diaria 

Las ganancias mensuales se calculan como $ r_{13}= log(\frac{p_3}{p_1}) $ 

Debido a la naturaleza aditiva de los logaritmos, se cumple que $r_{13}= r_{12} +r_{23}$

```python

assets["ret_oro"] = np.log(assets.oro/assets.lag_oro)
assets["ret_tesla"] = np.log(assets.tesla/assets.lag_tesla)
assets = assets.iloc[1:,] #eliminamos el primer mes para prescindir de los indeterminados (nans)
assets
```

```python
oro_nans =  assets.oro.isnull().values
tesla_nans= assets.tesla.isnull().values
oro_nans= len(oro_nans[oro_nans==True])
tesla_nans =  len(tesla_nans[tesla_nans==True])
{"indeterminados de Oro": oro_nans, "indeterminados de Tesla":tesla_nans }
```

```python
med_tesla= np.mean(assets.ret_tesla)
med_oro = np.mean(assets.ret_oro)
dt_tesla =np.std(assets.ret_tesla)
dt_oro = np.std(assets.ret_oro)
cov = np.cov(assets.ret_oro,assets.ret_oro)[0][0]
```

```python
print ({"activo": "acciones de tesla","retorno esperado": med_tesla, "volatilidad":dt_tesla})
print ({"activo": "oro","retorno esperado": med_oro, "volatilidad":dt_oro })
print(f"correlatividad de los activos:c{cov}")
```

rentabilidad=$ w_1E[R_1]+ w_2E[R_2]...$ donde $r_i$ es la rentabilidad activo i y $w_i$ la proporción de activos i en la cartera

volatilidad= $\sum \sum w_i*w_j *cov(r_i,r_j)$ donde $cov(r_i,r_j$ es la covarianza de los rendimientos de los activos i y j

```python
size = 100
W1 = list()  
W2 = list()
rents= list()
volats= list()
for i in range(size):
    weights = np.random.random(2)
    w1 = weights[0]/np.sum(weights)
    w2 = weights[1]/np.sum(weights)
    W1.append(w1)
    W2.append(w2)
    rents.append( w1*med_tesla+w2*med_oro)
    volats.append( w1*w2*cov)


    
```

```python
pf = pd.DataFrame({"prop_tesla":W1,"prop_oro":W2 ,"rentabilidad":rents,"volatilidad":volats})
pf
```

```python
plt.subplot()
plt.scatter(volats,rents)
plt.ylabel("rentabilidad")
plt.xlabel("volatilidad")
```

```python
max_vol = pf[pf.volatilidad== max(pf.volatilidad)]
cut = max_vol.rentabilidad.iloc[0]
pf = pf[pf.rentabilidad >= cut]
pf
```

```python
plt.subplot()
plt.scatter(pf.volatilidad,pf.rentabilidad)
plt.ylabel("rentabilidad")
plt.xlabel("volatilidad")
```

ahora vamos a buscar la mejor cartera de entre las que hemos generado. Para ello tenemos que calcular la rentabilidad ajsutada al riesgo (En adelante sharpe). 


Se calcula como $sharpe = \frac{r_i-r_f}{\sigma_i}$ donde $r_i$ es la rentabilidad de la cartera y $\sigma_i$ la volatilidad. 

$r_f$ es la rentabilidad de un activo sin riesgo, aunque despreciaremos esto en nuestros cálculos

```python
pf["sharpe"] = pf.rentabilidad/pf.volatilidad #cálculo de la rentabilidad ajustada
pf
```

```python
max_sharpe = pf[pf.sharpe == max(pf.sharpe)]
optimal_portfolio = {"proporción acciones de Tesla":max_sharpe.prop_tesla,
                    "proporción de inversión en oro":max_sharpe.prop_oro,
                    "rentabilidad esperada":max_sharpe.rentabilidad,
                    "volatilidad asumida":max_sharpe.volatilidad,
                    "rentabilidad ajustada al riesgo":max_sharpe.sharpe}
optimal_portfolio
```

veamos cómo se hallaría una optimización de cartera con los multiplicadores de lagrange

$max \space h(w_1,w_2)= \frac{w_1r_1+w_2r_2}{w_1w_2cov}+\lambda(w_1+w_2-1)$

$\frac{\partial h}{\partial w_1}= \frac{r_1(w_1w_2cov)-(w_1r_1+w_2r_2)w_2cov}{(w_1w_2cov)^2} +\lambda=0$

$\frac{\partial h}{\partial w_2}= \frac{r_2(w_1w_2cov)-(w_1r_1+w_2r_2)w_1cov}{(w_1w_2cov)^2}+\lambda=0$

$\frac{\partial h}{\partial \lambda} = w_1 +w_2 -1= 0 $

despejamos

$ \lambda = \frac{r_1}{w_2^2cov}$

$\lambda = \frac{r_2}{w_1^2cov}$

$w_1+ w_2 = 1$

Resolvemos

$\frac{r_1}{w_2^2cov} =  \frac{r_2}{w_1^2cov} \rightarrow w_1^2r_1 = w_2^2r_2= (1-w_1)^2r_2=(1-2w_1+w_1^2)r_2 \rightarrow w_1^2(r_1-r_2)+2r_2w_1-r_2=0$

$w_2=1-w_1$

$ax^2+bx+c=0 \rightarrow x= \frac{-b \pm \sqrt{b^2-4ac}}{2a}$

```python
a= med_tesla-med_oro
b=  2*med_oro
c=-med_oro
w_tesla = (-b+(b**2-4*a*c)**0.5)/(2*a)
w_tesla
```

```python
w_oro = 1- w_tesla
sharpe_lagrange = (w_tesla*med_tesla+w_oro*med_oro)/(w_tesla*w_oro*cov)

{"proporción óptima de acciones de tesla":w_tesla, "proporción óptima de activos en oro":w_oro, 
 "sharpe":sharpe_lagrange}
```
